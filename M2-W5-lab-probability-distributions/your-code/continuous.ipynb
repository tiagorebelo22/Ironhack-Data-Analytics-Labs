## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
import numpy as np
import matplotlib.pyplot as plt

def uniform_random_numbers(bottom, ceiling, count):
    x = uniform.rvs(size=count)
    a = bottom
    b = ceiling
    randoms  = a + (b-a)*x
    return np.array(randoms)


plot_options, (chart_1, chart_2) = plt.subplots(nrows=1, ncols=2)
chart_1.hist(uniform_random_numbers(10,15,100), bins=10)
chart_1.set_ylim(0,120)
chart_2.hist(uniform_random_numbers(10,60,1000), bins=10)
chart_2.set_ylim(0,120)

plt.show()
```

How are the two distributions different?

```python
"""
Both distributions are uniform. 
The first distribuition is ranged bewteen 10 and 15 and has only 100 values (smaller bars).
The second distribuition is ranged between 10 and 60 and has 1000 values (higher bars).
"""
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
def normal_random_numbers(mean, std, count):
    randoms = mean + std * np.random.randn(count)
    return np.array(randoms)

plot_options, (chart_1, chart_2) = plt.subplots(nrows=1, ncols=2)
chart_1.hist(normal_random_numbers(10,1,1000), bins=50)
chart_1.set_ylim(0,65)
chart_2.hist(normal_random_numbers(10,50,1000), bins=50)
chart_2.set_ylim(0,65)

plt.show()
```

How are the two distributions different?

```python
"""
Both distributions are normal. 
The first distribuition has a lower standard deviation (1), which is reflected by the narrower range of values.
The second distribuition has a higher standard deviation (50), which is reflected by the wider range of values.
"""
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with a mean of 1 and 100 and a size of 1000.

1. Plot the distributions as histograms with the number of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python

def exponential_random_numbers(mean, count):
    randoms = np.random.exponential(mean, size=count)
    return np.array(randoms)

plot_options, (chart_1, chart_2) = plt.subplots(nrows=1, ncols=2)
chart_1.hist(exponential_random_numbers(1,1000), bins=100)
chart_1.set_ylim(0,80)
chart_2.hist(exponential_random_numbers(100,1000), bins=100)
chart_2.set_ylim(0,80)

plt.show()
```

How are the two distributions different?

```python
"""
Both distributions are exponential.
The first distribuition has a lower mean (1), which is reflected by the narrower range of values (0 to around 8).
The second distribuition has a higher mean (100), which is reflected by the wider range of values (0 to around 700).
"""
```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# Hint: This is same as saying P(x<15)

from scipy.stats import expon

lambda_inv = 10
exp_dist = expon(scale = lambda_inv)

prob = exp_dist.cdf(14)

print("The probability that a customer will spend less than fifteen minutes in the bank: ", round(prob,3))
```

What is the probability that the customer will spend more than 15 minutes

```python
prob = 1-exp_dist.cdf(15)

print("The probability that the customer will spend more than 15 minutes: ",round(prob,3))
```
