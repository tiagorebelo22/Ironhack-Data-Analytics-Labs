```python
#Challenge 2: Count Commits

import datetime

url = "https://api.github.com/repos/tiagorebelo22/lab-list-comprehensions/commits"

# changed 2 weeks ago to 1 month ago so that the number of commits was not zero

two_weeks_ago = datetime.datetime.today() - datetime.timedelta(days=30)
two_weeks_ago = two_weeks_ago.strftime("%Y-%m-%dT%H:%M:%SZ")

response = requests.get(url,params={"since":two_weeks_ago})
response
```
