# StackAPI

#### Import the necessary libraries here:

```python
from stackapi import StackAPI
import pandas as pd
import datetime
```

#### Question 1: Find the questions and answers of last month.

```python
init_date = datetime.datetime.today().replace(day=1, hour=0, minute=0, second=0, microsecond=0)-datetime.timedelta(days=30)
end_date = datetime.datetime.today().replace(day=1, hour=0, minute=0, second=0, microsecond=0) - datetime.timedelta(seconds=1)

SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=int(init_date.timestamp()), todate=int(end_date.timestamp()))
questions
```

```python
answers = SITE.fetch('answers', fromdate=int(init_date.timestamp()), todate=int(end_date.timestamp()))
answers
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
init_date = datetime.datetime.today().replace(hour=0, minute=0, second=0, microsecond=0)
end_date = datetime.datetime.now()

questions = SITE.fetch('questions', fromdate=int(init_date.timestamp()), todate=int(end_date.timestamp()), min=5, tagged='python', sort='votes',order="desc")
questions

# No questions meet such criteria
```
